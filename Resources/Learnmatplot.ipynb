####Subplot and Legends

```python
days=[1,2,3,4,5]
original_prices=[10,22,30,42,50]
predicted_prices=[8,18,30,39,48]
```

```python
plt.plot(days, original_prices,'b-o',label='original prices');
plt.plot(days, predicted_prices,'r-o', label='predicted prices');
plt.legend()
```
